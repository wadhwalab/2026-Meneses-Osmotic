```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from pathlib import Path
from scipy.io import loadmat

# Ensure plots are rendered in the notebook
%matplotlib inline
```

```python
osm = "300mM"
```

```python
# Load the time series data from a .mat file using a relative path
file_path = os.path.join('.' , osm, 'cells\cell_2\cell_2_300_freq.mat')  # data for various osmotic shock values
fig_path = os.path.join('.' ,osm, 'example.pdf')  # location for saving plots
mat_data = loadmat(file_path)

# Extract the relevant data
# allcells = mat_data['AllCells']  # data for individual cells
speed = mat_data['F'].flatten()  # average across cells
# N = mat_data['N'].flatten()  # Number of cells
time = mat_data['T'].flatten()  # time


# cell = allcells[:,1]

# Plot the entire time series

# setting font sizeto 30
plt.rcParams.update({'font.size': 20})


plt.figure(figsize=(10, 5))
plt.plot(time, -speed, color = 'black', linewidth=0.75)
# plt.fill_between(time, lower, upper, color='gray', alpha = 0.2)
# plt.title(osm)
plt.xlabel('Time (s)')
plt.ylabel('Motor speed (Hz)')
plt.xlim(0,400)
plt.ylim(0,100)
plt.tick_params(which='both',direction='in')
plt.savefig(fig_path)
plt.show()
```
